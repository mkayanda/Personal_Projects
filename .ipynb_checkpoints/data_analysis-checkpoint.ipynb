### reducing memory by changing data types


```python
different_cols = [
         "RELAFFIL",
         "SATMTMID",
         "CURROPER",
         "INSTNM",
         "STABBR",
]
```

```python
col2 = college.loc[:, different_cols]
col2.head()
```

```python
#inspect the data types of each column
col2.dtypes
```

```python
#find the memory usage of each column 
original_mem = col2.memory_usage(deep=True)
original_mem
```

```python
#we can reduce size by
col2["RELAFFIL"] = col2["RELAFFIL"].astype(np.int8)
```

```python
# confirm if this change was made
col2.dtypes
```

```python
#find the moenety usdage of each column again
col2.memory_usage(deep=True)
```

```python
# we can save more memory by changing data types to categorical
col2.select_dtypes(include=["object"]).nunique()
```

```python
# this is a good candidate to convert to categoraical 
col2["STABBR"] = col2["STABBR"].astype("category")
```

```python
# compute the memory again
new_mem = col2.memory_usage(deep=True)
new_mem
```

```python
#finally lets compare original memry usage with our updated memery usage.

new_mem / original_mem
```
